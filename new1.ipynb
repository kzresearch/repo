```python
import csv
import numpy as np
import math as m
from matplotlib import pyplot

h = 6.626*10**-34 #Planck constant
c = 2.998*10**8 #speed of light
k_b = 1.381*10**-23 #Boltzmann constant
sigma = 5.670*10**-8 #Stefan-Boltzmann constant

#Function that reads a csv file and writes it to a matrix 
def csv_reader(input_file):
    
    reader = csv.reader(open(input_file, "rb"), delimiter=",")
    a = list(reader)        
    a_list = np.array(a).astype('float')
    return(a_list)
            
e = csv_reader("TER_file_cryorun7_emiss.csv") #emissivity matrix
d_r = csv_reader("TER_file_cryorun7_refl.csv") #diffuse reflectance matrix
s_r = csv_reader("specular_reflectance.csv") #specular reflectance matrix
t = csv_reader("TER_file_cryorun7_trans.csv") #transmittance matrix
```

```python
n_e = len(e[0]) #number of elements
n_k = len(e) #number of different wavenumbers
view_factor = 1 
step_size = 100 #increment size of wave number in m**-1 
t_0 = 300 #temperature of first element (index = 0)
t_n = 4.2 #temperature of last element (index = n_e -1)

E = [] #list for emissivity values
k_values = [] #list for wave numbers

#if loop to avoid divide by zero errors
for a in range(0, 400):
    for b in range(0, n_e):
        while (t[a,b] == 0):
            t[a,b]=10**-4
        while (d_r[a,b] == 0):
            d_r[a,b]=10**-4    

#Function to calculate voltage of first and last elements
#takes two arguemnts, wave number (k) and element temperature (T)
def volt(k, T):
    index = h*c*k/k_b/T
    if (index) > 10**2.8:
        return (0.)
    else:
        return (2.*h*c*k**3)*(1./((m.e**(h*c*k/k_b/T))-1.))
    
#Function to calculate divided current that goes through an element
#Takes 3 arguments: undivided current and index numbers to find required emissivity, reflectance and transmittance 
def I_2(current, a, b):
    return current*(1-t[a,b])*(1-s_r[a,b])*e[a,b]/(2*d_r[a,b]*t[a,b]+(1-t[a,b])*(1-s_r[a,b])*e[a,b])
                
            
#for loop to iterate over the whole wavenumber range (100-600000 m**-1)
for i in range(0, 400):
    
    k += step_size #increasing wave number by step size (100)
    
    '''Calculating total resistance (R_T) in order to find current'''
    
    #Total resistance between the first element (index=0) and the radiative field of 2nd element (index =1).
    R_T = d_r[i, 0]/((1-s_r[i, 0])*e[i, 0]) + 1/((1-s_r[i, 0])*(1-s_r[i, 1])*view_factor)
    
    #for loop to iterate over the rest of the elements excluding the last element
    for j in range(1, n_e-1):
        R_T += 2*d_r[i,j]*(1-t[i,j])/((1-s_r[i,j])*e[i,j]*(1-t[i,j])+2*d_r[i,j]*t[i,j]) +\
                1/((1-s_r[i, j])*(1-s_r[i, j+1])*view_factor)
    
    #final value of total resistance
    R_T += d_r[i, j+1]/(1-s_r[i, j+1])/e[i, j+1]

    ''' Calculating current'''
    #Current    
    I = (volt(k,t_0)-volt(k,t_n))/R_T 
    E_elements = [] #list of emissive power of elements
    E_0 = volt(k, t_0) #emissive power (potential) of first element (index = 0)
    E_elements.append(E_0)
    
    '''Calculating cumulative resistance in order to find emissive power of all the elements one by one'''
    #R_common is a set of resistors in series through which main current (I) passes
    #R_separate is in series with R_common but parallel with another set of resistors
    R_common = d_r[i, 0]/((1-s_r[i, 0])*e[i, 0]) + 1/((1-s_r[i, 0])*(1-s_r[i, 1])*view_factor)
    R_separate = d_r[i, 1]/((1-s_r[i, 1])*e[i, 1])
    
    #main current (I) divides into I_2 (calculated using I_2 function) before passing thorugh R_separate  
    #so, the potential of an element is difference of E_0 and products of I, R_common and I_2, R_separate  
    E_1 = E_0 - (I*R_common + I_2(I, i, 1)*R_separate)
    E_elements.append(E_1)
        
    #for loop to iterate over the rest of the elements except the last one
    for j in range(1, n_e-2):
        R_common += (2*d_r[i,j]*(1-t[i,j])/((1-s_r[i,j])*e[i,j]*(1-t[i,j])+2*d_r[i,j]*t[i,j])+
        1/(1-s_r[i,j])/(1-s_r[i,j+1])/view_factor)
        R_separate = d_r[i,j+1]/(1-s_r[i,j+1])/e[i,j+1]
        voltage = E_0 - (I*R_common + I_2(I, i, j+1)*R_separate)
        E_elements.append(voltage)
    
    j +=1
    R_common += (2*d_r[i,j]*(1-t[i,j])/((1-s_r[i,j])*e[i,j]*(1-t[i,j])+2*d_r[i,j]*t[i,j])+
        1/(1-s_r[i,j])/(1-s_r[i,j+1])/view_factor) + d_r[i,j+1]/(1-s_r[i,j+1])/e[i,j+1]
    
    E_elements.append(E_0 -I*R_common)
    
    E.append(E_elements) #adding lists of potentials of all elements into another list for varying wavenumber
    A = np.array(E) #converting the lists of potentials of all elements of different wavelengths to a numpy array
    k_values.append(k) #corresponding wavenumbers
    
    #print E_0, volt(k, t_n), I, R_T, R_common, E_0-I*R_T
    
    
```

```python
Q_T = [] #list of total energy of an element
T = [] #list of temperature of all the elements

#function that graphs Emissive power vs log of wavenumber for an element
#also outputs the area under the graph and temperature of the corresponding element
#takes one argument: element number
def plot(element_number):
    E_graph = [] #list of emissive power of an element for graphing
    for i in range(0, n_k):
        E_graph.append(A[i, element_number])
    
    print "plot for element no. %d" %(element_number+1)
    %matplotlib inline
    pyplot.semilogx(np.array(k_values), np.array(E_graph))
    bb = np.zeros(len(k_values))
    if element_number == 0:
        for j in range(len(k_values)):
            bb[j] += volt(float(k_values[j]),t_0)
        bb = bb/np.max(bb)*np.max(E_graph)
        pyplot.semilogx(np.array(k_values), bb,color='r')
    elif element_number == 12:
        for j in range(len(k_values)):
            bb[j] = volt(k_values[j],t_n)
        pyplot.semilogx(np.array(k_values), bb,color='r')
    pyplot.axis([100, 600000, np.amin(E_graph), np.amax(E_graph)])
    pyplot.show()
    
    print "Area under graph = %d" %(np.trapz(E_graph, k_values)*c*m.pi)
    print "Temperature of the element = %d" %((np.trapz(E_graph, k_values)*c*m.pi/sigma)**.25)
    Q_T.append(np.trapz(E_graph, k_values)*c*m.pi)
    T.append((np.trapz(E_graph, k_values)*c*m.pi/sigma)**.25)
```

```python
for i in range(0, n_e):
    plot(i)
```
